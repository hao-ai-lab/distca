```python
import pandas as pd
```

```python

dfw2 = pd.read_json('allgather.w2.jsonl', lines=True)
dfw4 = pd.read_json('allgather.w4.jsonl', lines=True)
```

```python
ax = dfw2.plot(x='tensor_size', y='mean_time_ms', kind='line', logx=True, logy=True, marker='o', label='world_size 2')
dfw4.plot(x='tensor_size', y='mean_time_ms', kind='line', logx=True, logy=True, marker='o', ax=ax, label='world_size 4')
ax.set_title('Allgather Latency')
ax.set_xlabel('Tensor Size (bytes)')
ax.set_ylabel('Mean Time (ms)')
ax.legend(title='World Size')
```
